```python
import requests

# Define the SPARQL query to retrieve crop pests and diseases entities from DBpedia
sparql_query = """
PREFIX dbo: <http://dbpedia.org/ontology/>
PREFIX dbr: <http://dbpedia.org/resource/>

SELECT ?entity
WHERE {
  ?entity a dbo:PlantDisease .
}
LIMIT 10
"""

# DBpedia SPARQL endpoint URL
endpoint_url = "http://dbpedia.org/sparql"

# Execute SPARQL query
response = requests.get(endpoint_url, params={'query': sparql_query, 'format': 'json'})
data = response.json()

# Extract entities from the query results
entities = [result['entity']['value'] for result in data['results']['bindings']]

# Print the extracted entities
print("Entities related to crop pests and diseases from DBpedia:")
for entity in entities:
    print(entity)
```

```python
import requests

# Define the SciGraph API URL and parameters
api_url = "https://scigraph.springernature.com/scigraph"
search_endpoint = "/terms/search"
params = {'prefix': 'crop pests and diseases', 'limit': 10}

# Execute the API request to search for entities related to crop pests and diseases
response = requests.get(api_url + search_endpoint, params=params)
data = response.json()

# Extract entities from the API response
entities = [term['label'] for term in data['results']]

# Print the extracted entities
print("Entities related to crop pests and diseases from SciGraph:")
for entity in entities:
    print(entity)
```
